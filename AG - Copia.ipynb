```python
import numpy as np
import random
```

```python
indi_size = 4 # 1 indivíduo -> 4 genes
popu_size = 6 # 1 populacao -> 6 indivíduos

def createPopulation(popu_size):
    chrom = np.random.randint(0,30, size=(popu_size,indi_size), dtype='int')
    return chrom

def evaluate(chrom):
    return (abs(chrom[0] + 2*chrom[1] + 3*chrom[2] + 4*chrom[3] - 30))

def fitness(f_obj):
    return (1/(1+f_obj))

def calcProb(ftss, total):
    return (ftss/total)

def calcCumProb(ftss,popu_size):
    cp = np.random.randint(0, size=(popu_size,), dtype='int')
    cp[0] = ftss[0]
    for i in range(1,popu_size):
        cp[i] = cp[i-1] + ftss[i]
    return cp

def rouletteWheel(popu_size):
    rw = np.random.random(popu_size,)
    return rw

def calcNewGeneration(chrom,rw,cumProb,popu_size): # rw = roulette wheel
    newchrom = np.zeros(shape=(6,4), dtype='int')
    for i in range(0,popu_size):
        for j in range(0,popu_size-1):
            if rw[i] > cumProb[j] and rw[i] < cumProb[j+1]:
                newchrom[i] = chrom[j+1]
                #print("troca:",i,"x",j+1)
            elif j==0 and rw[i] < cumProb[j]:
                #print("troca:",i,"x",0,"cond2")
                newchrom[i] = chrom[0]
    return newchrom

def crossOver(chrom,i,j,rnd):
    chrom[i][rnd:] = chrom[j][rnd:]
    
def mutation(chrom,position,value,popu_size):
    k = 1
    for i in range(0,popu_size):
        for j in range(0,4):
            if (k == position):
                chrom[i][j] = value
                return
            k += 1
```

```python
chrom = createPopulation(popu_size)
print("====== I N I C I O ======")
print(chrom)
print("======-------------======")
interator_number = 1
while True:
    f_obj = np.zeros(shape=(popu_size,), dtype='int')
    fitss = np.zeros(shape=(popu_size,), dtype='float')
    fitss_total = 0
    for i in range(0,popu_size):
        f_obj[i] = evaluate(chrom[i])
        if(f_obj[i] == 0):
            print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!",chrom[i])
        fitss[i] = fitness(f_obj[i])
        fitss_total += fitss[i]
    #print("f_obj:",f_obj)
    #print("fitss:",fitss)
    #print("fitss_total:",fitss_total)
    
    rnd = np.random.random(popu_size,)
    probChrMutation = np.zeros(shape=(popu_size,), dtype='float')
    probChrMutation_acm = np.zeros(shape=(popu_size,), dtype='float')
    for i in range(0,popu_size):
        probChrMutation[i] = calcProb(fitss[i],fitss_total)
        if (i==0):
            probChrMutation_acm[i] = probChrMutation[i]
        else:
            probChrMutation_acm[i] = probChrMutation_acm[i-1] + probChrMutation[i]
    
    #print("rnd_probMut:",rnd)
    #print("probChrMutation:",probChrMutation)
    #print("probChrMutation_acm:",probChrMutation_acm)
    
    newChrom = calcNewGeneration(chrom, rnd, probChrMutation_acm, popu_size)
    chrom = newChrom
    
    #print("bf_crossover:\n",chrom)
    
    rnd = np.random.random(popu_size,)
    for i in range(0,popu_size):
        if rnd[i] < 0.25:
            k = i+1
            if i>5:
                k = 0
            while(k < 6):
                if rnd[k] < 0.25:
                    crossOver(chrom,i,k,random.randint(1,3))
                    break
                k += 1
                if k == 6:
                    k = 0
    
    #print(rnd)
    #print("af_crossover:\n", chrom)
    
    total_gen = indi_size*popu_size
    mutation_rate = 0.25 # 10%

    for i in range(0,int(total_gen * mutation_rate)):
        position = random.randint(0,total_gen)
        mutation(chrom,position,random.randint(0,30),popu_size)
    
    print(chrom)
    print("================================")
    interator_number += 1
    if(interator_number > 100):
        break
print("======  F I N A L ======")
print(chrom)
for i in range(0,popu_size):
    print(evaluate(chrom[i]))
```
